## **Güven aralığı**
### *Güven aralığı*, istatistik biliminde bir anakütle parametresi için bir çeşit aralık kestirimi olup bir çıkarımsal istatistik çözüm aracıdır. Bir anakütle parametre değerinin tek bir sayı ile kestirimi yapılacağına, bu parametre değerini kapsayabilecek iki sayıdan oluşan bir aralık bulunur

# Örnek: 
### 0-80 arası 10.000 rastgele tamsayı üretelim. Bunun ortalaması ile bu anakitleden seçeceğimiz 100 rastgele sayının ortalamasının güven aralığındaki durumlarını inceleyelim.

```python
import numpy as np
np.random.seed(10)

anakütle=np.random.randint(0,80,10000) # 10.000 rastgele tamsayı

np.random.seed(10)
orneklem=np.random.choice(a=anakütle,size=100) # yukarıdaki örnekten rastgele seçeceğimiz 100 sayı örneklem
```

```python
orneklem.mean() # rastgele 100 sayının ortalaması
```

```python
anakütle.mean()
```

```python
import scipy.stats as st
```

```python
st.t.interval(alpha=0.95,
              df=len(orneklem)-1,
             loc=np.mean(orneklem),
              scale=st.sem(orneklem))
```

```python
import statsmodels.stats.api as sms
sms.DescrStatsW(orneklem).tconfint_mean()
```

### Örneklem ortalaması Anatükle ortalaması %95 güven aralığı içerisinde çıkmıştır.
